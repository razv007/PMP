```python
import numpy as np
import matplotlib.pyplot as plt
n = 1000
nums = [1,2,5,10]
X1 = np.random.poisson(1, n)
X2 = np.random.poisson(2, n)
X3 = np.random.poisson(5, n)
X4 = np.random.poisson(10, n)

probs=[0.1, 0.1, 0.7, 0.1]
alegeri = np.random.choice(nums, size=n, p=probs)
XR = np.array([np.random.poisson(l) for l in alegeri])

max_val=max(X1.max(), X2.max(), X3.max(), X4.max(), XR.max())
bins=np.arange(0, max_val+2)

fig, axes = plt.subplots(2, 3, figsize=(12, 6))
ax1 = axes[0, 0]
ax2 = axes[0, 1]
ax3 = axes[0, 2]
ax4 = axes[1, 0]
ax5 = axes[1, 1]
ax6 = axes[1, 2]

ax1.hist(X1, bins=bins, density=True, align='left')
ax1.set_title('Poisson(1)')
ax1.set_xlabel('Numbers')
ax1.set_ylabel('Densitate')
ax1.set_xlim(0, max_val)

ax2.hist(X2, bins=bins, density=True, align='left')
ax2.set_title('Poisson(2)')
ax2.set_xlabel('Numbers')
ax2.set_ylabel('Densitate')
ax2.set_xlim(0, max_val)

ax3.hist(X3, bins=bins, density=True, align='left')
ax3.set_title('Poisson(5)')
ax3.set_xlabel('Numbers')
ax3.set_ylabel('Densitate')
ax3.set_xlim(0, max_val)

ax4.hist(X4, bins=bins, density=True, align='left')
ax4.set_title('Poisson(10)')
ax4.set_xlabel('Numbers')
ax4.set_ylabel('Densitate')
ax4.set_xlim(0, max_val)

ax5.hist(XR, bins=bins, density=True, align='left')
ax5.set_title('Random')
ax5.set_xlabel('Numbers')
ax5.set_ylabel('Densitate')
ax5.set_xlim(0, max_val)

plt.tight_layout()
plt.show()
```
